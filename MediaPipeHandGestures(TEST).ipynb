```python
import mediapipe as mp
import os
import cv2
import uuid
import numpy as np
```

```python
mp_drawing = mp.solutions.drawing_utils
mp_hand = mp.solutions.hands
```

```python
cam = cv2.VideoCapture(0)

with mp_hand.Holistic(min_detection_confidence=0.8,min_tracking_confidence=0.5) as hands:
    while cam.isOpened():
        ret, frame = cam.read()
        image = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
        image.flags.writeable = False
        results = hands.process(image)
        image.flags.writeable = True
        image = cv2.cvtColor(frame, cv2.COLOR_RGB2BGR)
        print(results)

        mp_drawing.draw_landmarks(image, hands, mp_hand.HAND_CONNECTIONS)
                
        
        cv2.imshow("Hand Gestures", image)
        
        if cv2.waitKey(1) & 0xFF == ord('q'):
            break

cam.release()
cv2.destroyAllWindows()
```

```python
cam = cv2.VideoCapture(0)

with mp_hand.Hands(min_detection_confidence=0.8,min_tracking_confidence=0.5) as hands:
    while cam.isOpened():
        ret, frame = cam.read()
        image = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
        image.flags.writeable = False
        results = hands.process(image)
        image.flags.writeable = True
        image = cv2.cvtColor(frame, cv2.COLOR_RGB2BGR)
        print(results)

        if results.multi_hand_landmarks:
            for num, hand in enumerate(results.multi_hand_landmarks):
                mp_drawing.draw_landmarks(image, hand, mp_hand.HAND_CONNECTIONS)
        
        cv2.imshow("Hand Gestures", image)
        
        if cv2.waitKey(1) & 0xFF == ord('q'):
            break

cam.release()
cv2.destroyAllWindows()
```
